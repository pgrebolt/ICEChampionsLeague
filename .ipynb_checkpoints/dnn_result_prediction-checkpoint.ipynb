En aquest codi d'exemple veurem com carregar dades d'un fitxer csv i analitzar-les amb una xarxa neuronal. En aquest problema tenim un conjunt de dades mèdiques de diferents pacients i si tenen diabetis o no. Crearem el nostre model que ens permeti predir si un nou pacient tindrà diabetis o no en funció dels paràmteres d'entrada mesurats. És un problema de classificació.

Primer carreguem les dades i les separem en una matriu de dades, `X`, i un vector de resultats per cada conjunt de mesures, `y`.



```python
# first neural network with keras tutorial
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# load the dataset
dataset = loadtxt('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv', delimiter=',')
```

```python
# split into input (X) and output (y) variables
X = dataset[:,0:8] # hi ha 8 valors d'input
y = dataset[:,8] # ens diu si el pacient té o no té diabetis
```

Seguidament definim la nostra xarxa neuronal. És una xarxa seqüencial de capes denses. La primera és fully connected. Això vol dir que tots els inputs es tracten a totes les neurones de la primera capa. Amb la resta anem afegint capes. Aquest procés de tria de capes és assaig-error.

La funció d'activació RELU assigna un 0 si el resultat després d'ajustar els pesos és negatius i el deixa igual si aquest resultat és positiu. Es posa una sigmoide com a funció d'activació a la sortida. Això és per poder interpretar la sortida com una probabilitat. No és una probabilitat, però l'interpretem com a tal. La sigmoide el que fa és retornar un valor entre 0 i 1. Ho fem en aquest cas perquè és un cas de classificació (Diabètic/No diabètic). Ens interessa que la funció d'activació de l'última capa sigui diferenciable per poder fer després les optimitzacions.

```python
# define the keras model
model = Sequential() # el model serà una seqüència de capes
model.add(Dense(12, input_shape=(8,), activation='relu')) # 12 neurones. Es prenen els 8 valors d'input initials. És una fully connected
model.add(Dense(8, activation='relu')) # capa intermèdia de 8 neurones
model.add(Dense(1, activation='sigmoid')) #capa de sortida té només 1 neurona perquè és un problema de classificació (Sí/No)
```

```python
from keras.utils import plot_model

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

Compilem el model. Això implica definir la funció de les pèrdues i l'optimitzador. Són els paràmetres que es faran servir per optimitzar la xarxa en global. L'optimitzador `adam` té un algoritme intern que adjusta la learning rate automàticament. Aquest paràmetre és aquell que diu com canvia la predicció dels nous pesos amb un nou pas d'optimització.

Cal tenir en compte que quan es fa una optimització el que s'optimitza no és el que indica la mètrica (l'accuracy) sinó la funció de pèrdues. En el nostre cas és la `binary_crossentropy`.


```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Entrenem el model. L'època és cadascuna de les iteracions que es fa en l'entrenament. Es divideix la mostra d'entrenament en grups de mida definida pel `batch_size`. Els punts que van a cada batch es decideix de forma aleatòria. Aleshores es fa un entrenament i s'optimitza la xarxa per aquell grup. Després es passa al següent grup. Es fan tantes repeticions per tots els batchs com indiqui el paràmetres `epochs`.

Per escollir els valors d'`epochs` i `batch_size` definitius cal recórrer a la prova-error. Ens interessa un `batch_size` relativament petit perquè això indica que actualitzem els pesos del model amb més freqüència. Ara bé, si és massa petit el model pot no actualitzar-se correctament. Un `batch_size` petit també va bé per optimització de la memòria, ja que només s'hi carrega el número d'elements (mesures) determinat per aquest valor.

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
```

```python
# Avaluem el model
_, accuracy = model.evaluate(X, y)
print("Accuracy: %.2f" % (accuracy * 100))
```

```python
# Visualitzem els percentatges associats a cada input. Els interpretem com a probabilitat
model.predict(X).flatten()
```

```python
# Treiem la predicció per cada cas d'input
predictions = (model.predict(X) > 0.5).astype(int)
```

```python
# Visualitzem les prediccions i els valors reals de la dicotomia
import pandas as pd
pd.DataFrame({'pred':predictions.flatten(), 'real':y.astype('int')})
```

Hem vist que entrenar un model és relativament senzill. El que té importància és avaluar els resultats que ens torna. Per fer-ho, utilitzem diferents mètriques. L'avaluació amb diferents mètriques la fem amb `sci-kit learn`. Aquestes mètriques són necessàries per fer una bona avaluació del model.

L'accuracy depèn, en certa manera, de la representació que té cada classe a l'input de l'entrenament. Si una de les classes està sobrerepresentada pot canviar el valor de l'accuracy i fer-nos pensar que és bo, però en realitat tenim un model dolent. Calen les noves mètriques. Són la precisió i el recall. Tots aquests valors es poden trobar a partir de la confusion matrix.
- La precisió indica quants encerts espero tenir a partir de la decisió. Si el model diu que tinc 141+55 positius, però només 141 són correctes, la meva precisió és 0.72.
- El recall indica quants positius detecta el meu model. Si jo li dono 127+141 valors positius, però només en prediu 141 correctament, aleshores el meu recall és 141/(127+141) = 0.53.

Aquestes dues mètriques són complementàries i s'han de tenir en compte de forma conjunta. Es defineixen una per a cada classe. La discussió la fem respecte els positius (diabètics). Si tenim, per exemple, una precisió de 0.5 i un recall de 0.9 ens indica que si ens indica que la predicció és un 1, que sigui cert o fals és un 50%. Ara bé, com que el recall és alt ens indica que un 90% dels que sí que són diabètics els detecta. Amb un recall alt ens assegurem diagnosticar pràcticament tots els que tinguin diabetis. Si la precisió és baixa aleshores diagnosticarem com a diabètics a gent que no ho és. Val la pena? Depèn del cas. Aquests valors poden canviar si el threshold a `model.predict(X) > 0.5` puja o baixa. Ens quedem amb un threshold o un altre en funció del cas que estiguem tractant.

Per altra banda, tenim l'F1-score. És una mitjana geomètrica de la precisió i el recall. S'utilitza per tenir una espècie de ponderació respecte aquests dos paràmetres. Aquesta mètrica i l'AUC són mètriques utilitzades per comparar models diferents.

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

print(classification_report(y, predictions))
```

La matriu de confusió ens indica quants encerts (diagonal principal) i quants errors (diagonal secundària) tenim. A la diagonal principal hi trobem els True Positives i els True Negatives. En canvi, a la diagonal secundària hi tenim els False Positives i els False Negatives.

```python
ConfusionMatrixDisplay.from_predictions(y, predictions)
```

Una de les coses que no hem fet és dividir la nostra mostra inicial en una mostra d'entrenament i una altra de test. Ho farem seguidament per tornar a entrenar el nostre model i per avaluar-lo. Ho fem amb `sci-kit learn`. Aquesta divisió de la mostra ens permet catalogar com es comporta el nostre model per predir dades noves. Les dades que s'assignen com a text no es fan servir en l'entrenament. Quan avaluem el model, actuen com si fossin dades noves. L'avaluació de les prediccions es fa sobre les dades test, mentre que l'entrenament es fa sobre les dades train.

```python
from sklearn.model_selection import train_test_split

# dividim la mostra, per defecte la ratio és 75:25
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Fem el fit del model (definit abans) a les mostres train
model.fit(X_train, y_train, epochs=150, batch_size=10)

# Calculem les prediccions per les mostres test. Assignem 1 a les que tenen una predicció més gran que 0.5
y_pred = (model.predict(X_test) > 0.5).astype(int).flatten()

# Mostrem els resultats, en base a les mostres test.
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

Una altra cosa que se sol fer és escalar les dades abans d'introduir-les al model. Es pren cada columna i, a cada valor, se li resta la mitjana de la columna i es divideix per la desvest de la columna. Així ens assegurem que els valors que hi ha cada columna tenen mitjana 0 i desvest 1. Això ens permet treure grans canvis de valors que hi pugui haver en cada paràmetre del model (cada paràmetre del model es troba anotat en una columna). Aquest escalat es fa amb `skit-learn`.

Aquest escalat pot facilitar la predicció de la xarxa. És un pas que forma part del propi model! També forma part del model el pas previ de tractament de dades en el sentit d'emplenar o descartar mesures on hi hagi valors que hi faltin. En general, els passos a seguir són:

Carregar dades - Feature engeneering (extracció de paràmetres de les dades incials; depèn de l'àmbit al qual s'apliqui el model; p.ex. determinar l'amplada d'un pic d'un espectre) - Preprocessat (emplenar NAs, escalar dades) - Model - Predicció

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # inicialitzem l'escalador
X_train_sc = scaler.fit_transform(X_train) # calculem la mitjana i la desvest a cada columna i aplqiuem la transformació, per train
X_test_sc = scaler.transform(X_test) # apliquem la transformació a test (paràmetres calculats per train)
```

```python
# Fem el fit del model (definit abans) a les mostres train
model.fit(X_train_sc, y_train, epochs=150, batch_size=10)

# Calculem les prediccions per les mostres test. Assignem 1 a les que tenen una predicció més gran que 0.5
y_pred = (model.predict(X_test_sc) > 0.5).astype(int).flatten()

# Mostrem els resultats, en base a les mostres test.
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```
